This script is used to obtain the dynids and trajids that are used in the project. It is very important to compute it correctly, as there are pockets calculated from non-anymore-existing simulations, duplicated pockets, etc. So I have to take them from the `compl_info.json` file as they are curated there and cross-validate them with the files in `mdpocket_oversized`.

```python
import os
import json
```

```python
# Paths to run locally.
PROJECT_ROOT = "/home/alex/sshfs_mountpoints/verde/Documents/pocket_tool"
ROOT_GPCRMD = "/home/alex/sshfs_mountpoints/gpcrmd"
pocket_dir = "/home/alex/sshfs_mountpoints/mdpocket_oversized/"

COMPL_INFO_PATH = os.path.join(ROOT_GPCRMD, "media/files/Precomputed/compl_info.json")
```

```python
def read_json_into_dict(json_path):
    """
    Reads a JSON file and returns the data as a dictionary.
    """
    with open(json_path, 'r') as f:
        data = json.load(f)
    return data


def get_trajetory_ids_from_dynid(dynid, compl_info):
    """
    Retrieves trajectory IDs associated with a given dynamic ID.
    """
    trajids = []
    traj_files = compl_info['dyn'+str(dynid)]['traj_f']
    for file in traj_files:
        basename = os.path.basename(file)
        trajid = basename.split("_")[0]
        trajids.append(trajid)
    return trajids
```

```python
compl_info = read_json_into_dict(COMPL_INFO_PATH)
```

```python
for compl_dynid_name in compl_info:
    compl_dynid = compl_dynid_name.replace('dyn', '').strip()

    traj_names = compl_info['dyn'+str(compl_dynid)]['traj_f']

    for traj_name in traj_names:
        traj_basename = os.path.basename(traj_name)
        mdpocket_dir_name = f'{traj_basename}_mdpocket'

        if os.path.isdir(os.path.join(pocket_dir, mdpocket_dir_name)):
            trajid = traj_basename.split("_")[0]
            print(f"{compl_dynid};{trajid}")
        else:
            print(f"[WARNING] dynamic ID {compl_dynid} & trajID {traj_basename} is missing mdpocket data.")
        
```
